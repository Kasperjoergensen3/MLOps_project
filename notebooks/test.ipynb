```python
import matplotlib.pyplot as plt
import torch
import numpy.random as random
import json


DATA_PATH = "../data/processed/train.pt"
JSON_PATH = "../data/processed/class_dictionary.json"

dataset = torch.load(DATA_PATH)
with open(JSON_PATH, "r") as f:
    class_dict = json.load(f)

#
N = len(dataset)
print(f"Dataset size: {N}")

# print shape of first image
print(f"Image shape: {dataset[0][0].shape}")
# get 16 random numbers
random.seed(0)
idx = random.randint(0, N, 16)

fig, axs = plt.subplots(4, 4, figsize=(10, 10))

axs = axs.flatten()
for i, ax in enumerate(axs):
    ax.imshow(dataset[idx[i]][0].numpy().squeeze(), cmap="gray")

    ax.set_title(class_dict[str(dataset[idx[i]][1].item())])
    ax.axis("off")

plt.savefig("../reports/figures/visualize_dataset.png")
```
